# **Age**

# **In this company , Maximum attrition happens whose  age 25-35 employees**

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(pd.crosstab(df['Age'],df['Attrition']),annot=True, fmt="d")
```

**Under our analysis Probability of attrition of leaving company to staying in company is relatively higher between 
age 20-35** 

```python

sns.distplot(df[df['Attrition']== 0]['Age'],hist=False)
sns.distplot(df[df['Attrition']== 1]['Age'],hist=False)
```

# **Business Travel**

**Under our Analysis Maximum attrition in the company is done by employess travelling rarely**

```python
sns.heatmap(pd.crosstab(df['BusinessTravel'],df['Attrition']),annot=True, fmt="d")
```

**Under our Analysis 0.25 is the highest probability of attrition in the company by employees given who travel Frequently**

```python
BusinessTravel_crosstab = pd.crosstab(df.Attrition,columns = df.BusinessTravel)
BusinessTravel_crosstab
rowtotal = BusinessTravel_crosstab.sum(axis=1)
coltotal = BusinessTravel_crosstab.sum(axis=0)
BusinessTravel_crosstab
given_Attrition_BusinessTravel = (BusinessTravel_crosstab.div(coltotal,axis=1))#*100
given_Attrition_BusinessTravel
sns.heatmap(given_Attrition_BusinessTravel,annot=True)
```
